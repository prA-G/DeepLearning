# *Perceptron*

A perceptron is the simplest neural network unit used for binary classification (yes/no, 0/1).

- It takes inputs
- multiplies them with weights
- adds a bias
- passes through an activation function
- gives an output

```python
# Import all libraries and framework
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating three data frame
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input 1'] = [1,1,0,0]
or_data['input 2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]

print(or_data)
```

```python
and_data['input 1'] = [1, 1, 0, 0]
and_data['input 2'] = [1, 0, 1, 0]
and_data['output'] = [1, 0, 0, 0]

print(and_data)
```

```python
xor_data['input 1'] = [1, 1, 0, 0]
xor_data['input 2'] = [1, 0, 1, 0]
xor_data['output'] = [0, 1, 1, 0]

print(xor_data)
```

```python
# Creating Visualization of all the above mentioned gates

sns.scatterplot(x=or_data['input 1'], y=or_data['input 2'], hue=or_data['output'], s = 200)
```

```python
sns.scatterplot(x=and_data['input 1'], y=and_data['input 2'], hue=and_data['output'], s = 200)
```

```python
sns.scatterplot(x=xor_data['input 1'], y=xor_data['input 2'], hue=xor_data['output'], s = 200)
```

```python
# Creating a perceptron classifier using sklearn

from sklearn.linear_model import Perceptron

classifier_or = Perceptron()
classifier_and = Perceptron()
classifier_xor = Perceptron()
```

```python
classifier_or.fit(or_data.iloc[:,0:2].values, or_data.iloc[:,-1].values)
classifier_and.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
classifier_xor.fit(xor_data.iloc[:,0:2].values, xor_data.iloc[:,-1].values)
```

## Limitations of Perceptron

- It is easily able to create the linear decision boundaries but for non-linear functions, It wont be able to create the linear decision boundary.

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# Set-up subplots for side by side comparison
fig, axes = plt.subplots(1, 3, figsize=(18,5))

# AND Gate classifier Decision Boundary
axes[0].set_title('AND Gate Decision Boundary')
plot_decision_regions(X=and_data.iloc[:, 0:2].values, y=and_data.iloc[:,-1].values, clf=classifier_and, legend= 2, ax=axes[0])
axes[0].set_xlabel('Input 1')
axes[0].set_ylabel('Input 2')

# OR Gate classifier Decision Boundary
axes[1].set_title('OR Gate Decision Boundary')
plot_decision_regions(X=or_data.iloc[:, 0:2].values, y=or_data.iloc[:,-1].values, clf=classifier_or, legend= 2, ax=axes[1])
axes[1].set_xlabel('Input 1')
axes[1].set_ylabel('Input 2')

# XOR Gate classifier Decision Boundary
axes[2].set_title('XOR Gate Decision Boundary')
plot_decision_regions(X=xor_data.iloc[:, 0:2].values, y=xor_data.iloc[:,-1].values, clf=classifier_xor, legend= 2, ax=axes[2])
axes[2].set_xlabel('Input 1')
axes[2].set_ylabel('Input 2')

plt.tight_layout()
plt.show()
```

- For AND and OR gate, Perceptron works well in linear decision boundary but for  XOR gate perceptron fails in making clear cut boundary.
